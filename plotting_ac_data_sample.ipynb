```python
import pickle

import numpy as np
import matplotlib.pyplot as plt
```

```python
with open(drive_path + 'arabchars.pkl', 'rb') as f:
    ac_data = pickle.load(f)
```

```python
train_imgs = ac_data[0].copy()
test_imgs = ac_data[1].copy()
train_labels = ac_data[2].copy()
test_labels = ac_data[3].copy()
X_train = ac_data[4].copy()
X_test = ac_data[5].copy()
y_train = ac_data[6].copy()
y_test = ac_data[7].copy()
class_dict = ac_data[8].copy()
```

```python
count=1
plt.subplots(figsize=(15, 15))
for i in np.random.randint(0, len(train_labels), 25):
   plt.subplot(5, 5, count)
   plt.title(label=f'{class_dict[train_labels[i]][0]}  -  {class_dict[train_labels[i]][1]}', fontsize=22)
   plt.imshow(train_imgs[i], cmap='Greys_r')
   count+=1
plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.show()

plt.savefig(drive_path + 'data_sample.png')
```
